1. What is a random variable in probability theory?

A random variable is a function that assigns a real number to each outcome of a random experiment. It quantifies outcomes and allows us to analyze them mathematically.

2. What are the types of random variables?

There are two main types:

Discrete Random Variable: Takes countable values (e.g., number of students).

Continuous Random Variable: Takes values within a range (e.g., height, weight).

3. What is the difference between discrete and continuous distributions?

Discrete Distributions describe probabilities of distinct, separate outcomes (e.g., number of heads in coin tosses).

Continuous Distributions describe probabilities over continuous intervals, where exact probabilities for single points are 0.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The CLT states that the sampling distribution of the sample mean becomes approximately normal as the sample size increases, regardless of the original distribution.

It's key for statistical inference using the normal distribution.

14. How does the Central Limit Theorem relate to the normal distribution?
The CLT ensures that sample means are approximately normally distributed when sample size is large, allowing us to apply normal-based methods like Z-tests and confidence intervals.

17. What are point estimates and interval estimates in statistics?
Point Estimate: A single value that estimates a parameter (e.g., sample mean).

Interval Estimate: A range of values that likely contains the population parameter (e.g., confidence interval).

18. What is the significance of confidence intervals in statistical analysis?
A confidence interval (CI) gives a range for a population parameter, along with a level of confidence (e.g., 95%) that the true parameter lies within that range.

```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, binom, poisson

# 1. Simulate multiple samples from a normal distribution and verify CLT
sample_means = []
for _ in range(1000):
    sample = np.random.normal(loc=50, scale=10, size=30)
    sample_means.append(np.mean(sample))
plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.title("<B5 Central Limit Theorem Verification")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid(True)
plt.show()

# 2. Plot standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.title("<*5 Standard Normal Distribution")
plt.xlabel("Z")
plt.ylabel("PDF")
plt.grid(True)
plt.show()

# 3. Generate binomial random variables and calculate probabilities
binomial_data = binom.rvs(n=10, p=0.5, size=1000)
prob = binom.pmf(5, n=10, p=0.5)
print("<E5 P(X=5) for Binomial(n=10, p=0.5):", prob)

# 4. Calculate Z-score and compare to standard normal distribution
x_value = 80
mean = 70
std_dev = 10
z_score = (x_value - mean) / std_dev
prob_standard = norm.cdf(z_score)
print("<5 Z-score:", z_score, "Cumulative probability:", prob_standard)

# 5. Hypothesis testing using Z-statistics
sample_mean = 72
population_mean = 70
std_dev = 10
n = 30
z = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 1 - norm.cdf(z)
print("<5 Z-statistic:", z, "P-value:", p_value)

# 6. Confidence interval for dataset
data = np.random.normal(100, 15, 50)
mean_data = np.mean(data)
std_error = np.std(data, ddof=1) / np.sqrt(len(data))
margin = 1.96 * std_error
conf_int = (mean_data - margin, mean_data + margin)
print("9\"5 95% Confidence Interval:", conf_int)

# 7. Confidence interval from generated normal data
sample = np.random.normal(loc=60, scale=8, size=40)
mean_s = np.mean(sample)
se = np.std(sample, ddof=1) / np.sqrt(len(sample))
margin_s = 1.96 * se
ci_s = (mean_s - margin_s, mean_s + margin_s)
print("9<5 Confidence Interval for Sample Mean:", ci_s)

# 8. PDF of a normal distribution
x_vals = np.linspace(60, 80, 100)
pdf_vals = norm.pdf(x_vals, loc=70, scale=5)
plt.plot(x_vals, pdf_vals)
plt.title("995 Normal Distribution PDF")
plt.xlabel("x")
plt.ylabel("PDF")
plt.grid(True)
plt.show()

# 9. CDF of Poisson distribution
cdf_poisson = poisson.cdf(np.arange(0, 10), mu=3)
plt.step(np.arange(0, 10), cdf_poisson, where='mid')
plt.title("9\x16  Poisson CDF (mu=3)")
plt.xlabel("k")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()

# 10. Continuous uniform variable and expected value
uniform_data = np.random.uniform(5, 15, 1000)
expected_val = np.mean(uniform_data)
print("9\x12 Expected Value of Uniform(5, 15):", expected_val)

# 11. Compare standard deviations of two datasets
set1 = np.random.normal(100, 5, 100)
set2 = np.random.normal(100, 15, 100)
std1 = np.std(set1)
std2 = np.std(set2)
plt.hist(set1, bins=20, alpha=0.5, label='Low SD')
plt.hist(set2, bins=20, alpha=0.5, label='High SD')
plt.title("9B5 Comparing Standard Deviations")
plt.legend()
plt.grid(True)
plt.show()
print("Standard Deviation Set1:", std1)
print("Standard Deviation Set2:", std2)

# 12. Range and IQR of a dataset
normal_sample = np.random.normal(50, 10, 100)
data_range = np.ptp(normal_sample)
q1 = np.percentile(normal_sample, 25)
q3 = np.percentile(normal_sample, 75)
iqr = q3 - q1
print("9*5 Range:", data_range, "IQR:", iqr)

# 13. Z-score normalization
z_normalized = (normal_sample - np.mean(normal_sample)) / np.std(normal_sample)
sns.histplot(z_normalized, kde=True)
plt.title("9E5 Z-score Normalization")
plt.xlabel("Z-score")
plt.grid(True)
plt.show()
```
